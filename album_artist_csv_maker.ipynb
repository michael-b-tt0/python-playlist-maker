```python
import pandas as pd
from pathlib import Path
import re

directory = Path(r"K:\sample music for playlist makers")

names = [p.name for p in directory.iterdir() if p.is_dir()]

rows = [
    re.split(
        r"\s*(?:—|\s-\s)\s*",
        re.split(r"[\(\[\{]", n, 1)[0].strip(),
        maxsplit=1
    )
    for n in names
]

df = pd.DataFrame(
    [
        (
            r[0].strip(),
            r[1].strip().rstrip("-–— ").strip() if len(r) > 1 else None
        )
        for r in rows
    ],
    columns=["artist", "title"]
)

print(df)
df.to_csv("albums.csv", index=False)


```
